```python
synergy_df = pd.read_csv('nba_offense.csv')
synergy_df.head()
```

## Process Data

### Compute Possession Count for Players

```python
positions = ['PG', 'SG', 'SF', 'PF', 'C']
```

```python
ctg_name_dict = {'Otto Porter': 'Otto Porter Jr.', 
                 'Michael Porter': 'Michael Porter Jr.',
                 'Kevin Porter': 'Kevin Porter Jr.',
                 'Kelly Oubre': 'Kelly Oubre Jr.',
                 'Jaren Jackson': 'Jaren Jackson Jr.',
                 'Wendell Carter': 'Wendell Carter Jr.',
                 'Troy Brown': 'Troy Brown Jr.',
                 'Chaundee Brown': 'Chaundee Brown Jr.',
                 'Duane Washington': 'Duane Washington Jr.',
                 'David Duke': 'David Duke Jr.',
                 'Charles Brown': 'Charlie Brown Jr.',
                 'Zach Norvell': 'Zach Norvell Jr.',
                 'RJ Nembhard': 'RJ Nembhard Jr.',
                 'Larry Drew': 'Larry Drew II',
                 'Brian Bowen': 'Brian Bowen II',
                 'Robert Williams': 'Robert Williams III',
                 'Marvin Bagley': 'Marvin Bagley III',
                 'Lindy Waters': 'Lindy Waters III',
                 'Greg Brown': 'Greg Brown III',
                 'Lonnie Walker': 'Lonnie Walker IV',
                 'Wade Baldwin': 'Wade Baldwin IV',
                 'PJ Washington': 'P.J. Washington',
                 'TJ McConnell': 'T.J. McConnell',
                 'TJ Leaf': 'T.J. Leaf',
                 'TJ Warren': 'T.J. Warren',
                 'Zach LaVine': 'Zach Lavine',                
                 'Cameron Thomas': 'Cam Thomas',
                 'Sviatoslav Mykhailiuk': 'Svi Mykhailiuk'}
```

```python
ctg_dfs = []
for pos in positions:
    temp = lineup_df[[pos, 'Season', 'team', 'Poss']].copy()
    temp = temp.rename(columns={'team': 'Team',
                                pos: 'Player'})
    ctg_dfs.append(temp)
ctg_df = pd.concat(ctg_dfs).reset_index(drop=True)\
       .groupby(['Season', 'Team', 'Player'])[['Poss']].sum()\
       .reset_index()\
       .rename(columns={'Poss': 'Off_Poss_On_Court',
                        'Player': 'Player_CTG'})
ctg_df['Player'] = ctg_df['Player_CTG']\
                   .apply(lambda x: ctg_name_dict.get(x, x))
ctg_df.head()
```

```python
ctg_df = ctg_df.merge(on_off_df[['Player_CTG', 'Season', 'Team',
                                 'off_Pts/Poss', 'def_Pts/Poss',
                                 'Diff']],
                      how='left',
                      on=['Player_CTG', 'Season', 'Team'])\
               .rename(columns={'off_Pts/Poss': 'onpm', 
                                'def_Pts/Poss': 'dnpm',
                                'Diff': 'npm'})
ctg_df.head()
```

### Merge

```python
df = synergy_df.merge(ctg_df,
                      on=['Season', 'Team', 'Player'],
                      how='left')
```

```python
df[df['Off_Poss_On_Court'].isna()]\
.sort_values('Overall_Poss', ascending=False)
```

```python
df = df[~df['Off_Poss_On_Court'].isna()].reset_index(drop=True)
```

### Create New Features

```python
situations = ['Pick and Rolls Including Passes',
              'Isolations Including Passes',
              'Post-Ups Including Passes']
```

```python
for situation in situations:
    df[f'{situation}_On_Court%'] = df[f'{situation}_Poss'] / \
                                   df['Off_Poss_On_Court']
    print(situation, 
          df[f'{situation}_On_Court%'].max(),
          df[f'{situation}_On_Court%'].min())
df.head()
```
